# Análise descritiva sobre ações da Petrobras

- Período dos dados: **03 de fev. de 2023 - 03 de mai. de 2023**
- Dados encontrados em: [Yahoo Finanças](https://br.financas.yahoo.com/quote/PBR/history?period1=1675468800&period2=1683158400&interval=1d&filter=history&frequency=1d&includeAdjustedClose=true)

```python
from google.colab import files
import io

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# Carregando arquivo - dados das ações

uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['PBR.csv']))
```

```python
# Estatísticas descritivas dos dados
df.describe()
```

```python
# Coeficiente de variação
CV = df.std()/df.mean() * 100
CV
```

```python
# Histogramas
nome = 'Close'
df[nome].hist()

plt.xlabel('Preço', size=14)
plt.ylabel('Frequência', size=14)

# Mediana e Média 
plt.plot(df[nome].median(), 0, '*r')
plt.plot(df[nome].mean(), 0, 'og')
plt.show()
```

```python
# Boxplot
df.iloc[:, 0:6].boxplot()
plt.show()
```
